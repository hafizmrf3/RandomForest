```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

import pandas as pd
import numpy as np
np.random.seed(0)
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
#tambahkan kolom baru untuk nama spesies
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.tail(70)
```

```python
df['is_train'] = np.random.uniform(0,1,len(df)) <= .75
df.head()
```

```python
#membagi dataframes dengan data latih dan data uji
train, test = df[df['is_train']==True], df[df['is_train']==False]
#tampilkan total pengamatan untuk dataframes latih dan uji
print("Total pengamatan pada Data Training", len(train))
print("Total pengamatan pada Data Uji", len(test))
```

```python
features = df.columns[:4]
features
```

```python
y = pd.factorize(train['species'])[0]
y 
```

```python
clf = RandomForestClassifier(n_jobs=2, random_state=0)

clf.fit(train[features], y)
```

```python
clf.predict(test[features])
```
